```python
#T1
import pandas as pd
netflix = pd.read_csv("netflix_titles.csv")
```

```python
#T2. Display the first 5 and last 5 rows
print("First 5 rows:")
print(netflix.head(5))
print("Last 5 rows:")
print(netflix.tail(5))
```

```python
#T3. How many rows and columns are there in the dataset?
print("Shape of the dataset (rows, columns):", netflix.shape)
```

```python
#T4. Identify the number of unique countries where content is produced
unique_countries = netflix['country'].nunique()
print("Number of unique countries:", unique_countries)
```

```python
#T5. Display all records where the type is 'TV Show'
tv_shows = netflix[netflix['type'] == 'TV Show']
print("All TV Shows:")
print(tv_shows)
```

```python
#T6. Convert the date_added column to datetime and extract the year and month
netflix['date_added'] = pd.to_datetime(netflix['date_added'])
netflix['year_added'] = netflix['date_added'].dt.year
netflix['month_added'] = netflix['date_added'].dt.month
print(netflix[['date_added', 'year_added', 'month_added']].head(3))
```

```python
#T7. How many movies were released in 2019?
movies_2019 = netflix[(netflix['type'] == 'Movie') & (netflix['release_year'] == 2019)]
print("Number of movies released in 2019:", movies_2019.shape[0])
```

```python
#T8. Group by rating and show the count of each category
rating_counts = netflix['rating'].value_counts()
print("Rating counts:")
print(rating_counts)
```

```python
#T1: Load the dataset and Display top 10 rows
world = pd.read_csv("world_population.csv")
world.head(10)
```

```python
# T2: Check for null values
world.isnull().sum()

#Handling
world_cleaned = world.dropna()

print("Original rows:", len(world))
print("Cleaned rows:", len(world_cleaned))
```

```python
#T3:Population of the top 5 most populous countries
top_5 = world_cleaned.sort_values(by='2020 Population', ascending=False)
top_5[['Country/Territory', '2020 Population']].head(5)
```

```python
#T4:Create GDP per capita column
import numpy as np
np.random.seed(0)
world_cleaned['GDP'] = np.random.randint(1000, 50000, size=len(world_cleaned))
world_cleaned['GDP per Capita'] = world_cleaned['GDP'] / world_cleaned['2020 Population']
world_cleaned[['Country/Territory', 'GDP per Capita']].head()
```

```python
#T5:Filter countries with population >100M and area <500,000 sq km
filtered = world_cleaned[
    (world_cleaned['2020 Population'] > 1e8) & 
    (world_cleaned['Area (km²)'] < 500000)
]
filtered[['Country/Territory', '2020 Population', 'Area (km²)']]
```

```python
#T6:Sort countries based on density (descending)
sorted_density = world_cleaned.sort_values(by='Density (per km²)', ascending=False)
sorted_density[['Country/Territory', 'Density (per km²)']].head()
```

```python
# T1: Load the datasets
matches = pd.read_csv("matches.csv")
deliveries = pd.read_csv("deliveries.csv")
```

```python
# T2: Find the most successful team based on number of wins
most_wins = matches['winner'].value_counts()
print("Most successful team:", most_wins.idxmax())
print("Total wins:", most_wins.max())
```

```python
# T3: Count total matches played each season
matches_per_season = matches['season'].value_counts().sort_index()
print("Matches per season:")
print(matches_per_season)
```

```python
# T4: Top 5 players with most 'Player of the Match' awards
top_players = matches['player_of_match'].value_counts().head(5)
print("Top 5 players with most Player of the Match awards:")
print(top_players)
```

```python
# T5: Merge matches.csv and deliveries.csv on match_id
merged_df = pd.merge(deliveries, matches, left_on='match_id', right_on='id')
merged_df.head()
```

```python
#Task 1: Average marks in math, reading, and writing
students = pd.read_csv("StudentsPerformance.csv")
print("Average Math Score:", students['math score'].mean())
print("Average Reading Score:", students['reading score'].mean())
print("Average Writing Score:", students['writing score'].mean())
```

```python
#T2: Who performed better in writing – males or females?
writing_by_gender = students.groupby('gender')['writing score'].mean()
print("Average writing scores by gender:")
print(writing_by_gender)

better = writing_by_gender.idxmax()
print(f"\n{better.capitalize()}s performed better in writing.")
```

```python
#T3: Add total score and percentage columns
students['total score'] = students['math score'] + students['reading score'] + students['writing score']
students['percentage'] = students['total score'] / 3
students[['math score', 'reading score', 'writing score', 'total score', 'percentage']].head()
```

```python
#T4: Bar chart of average scores by parental level of education
import matplotlib.pyplot as plt

avg_scores = students.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()
avg_scores.plot(kind='bar', figsize=(10,6))
plt.title("Average Scores by Parental Level of Education")
plt.ylabel("Average Score")
plt.xlabel("Parental Level of Education")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#T5: Find students who scored above 90%
top_students = students[students['percentage'] > 90]
print("Students who scored above 90%:")
top_students.head()
```
